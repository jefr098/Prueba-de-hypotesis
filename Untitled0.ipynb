```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_excel('tstudent_algoritmos.xlsx')

A = df['Algoritmo_A_ops_sec']
B = df['Algoritmo_B_ops_sec']

# 📊 Histogramas
plt.figure()
plt.hist(A, alpha=0.5, label='Algoritmo A')
plt.hist(B, alpha=0.5, label='Algoritmo B')
plt.legend()
plt.title("Histogramas - Algoritmos")
plt.show()

# 📊 Boxplot
plt.figure()
sns.boxplot(data=[A, B])
plt.xticks([0,1], ['A','B'])
plt.title("Boxplot - Algoritmos")
plt.show()

# Tests
print("Shapiro A:", stats.shapiro(A))
print("Shapiro B:", stats.shapiro(B))
print("Levene:", stats.levene(A, B))

t_stat, p_value = stats.ttest_ind(A, B, equal_var=False)
p_value_one_tail = p_value / 2

print("t:", t_stat)
print("p:", p_value_one_tail)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_excel('mannwhitney_redes.xlsx')

X = df['Red_X_ms']
Y = df['Red_Y_ms']

# 📊 Histogramas
plt.figure()
plt.hist(X, alpha=0.5, label='Red X')
plt.hist(Y, alpha=0.5, label='Red Y')
plt.legend()
plt.title("Latencia Redes")
plt.show()

# 📊 Boxplot
plt.figure()
sns.boxplot(data=[X, Y])
plt.xticks([0,1], ['X','Y'])
plt.title("Comparación de latencias")
plt.show()

# Test
u_stat, p_value = stats.mannwhitneyu(X, Y, alternative='less')

print("U:", u_stat)
print("p:", p_value)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

df = pd.read_excel('anova_cifrado.xlsx')

# 📊 Boxplot general
plt.figure()
sns.boxplot(data=df)
plt.title("Consumo CPU por algoritmo")
plt.show()

# 📊 Histogramas
df.hist(figsize=(8,5))
plt.show()

# ANOVA
f_stat, p_value = stats.f_oneway(df['AES_256'], df['RSA_2048'], df['ChaCha20'])

print("F:", f_stat)
print("p:", p_value)

if p_value < 0.05:
    print("Hay diferencias → Tukey")

    data = pd.melt(df)
    tukey = pairwise_tukeyhsd(data['value'], data['variable'])
    print(tukey)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_excel('correlacion_servidor.xlsx')

x = df['Solicitudes_Concurrentes']
y = df['Tiempo_Respuesta_ms']

# 📊 Scatter plot
plt.figure()
plt.scatter(x, y)
plt.xlabel("Solicitudes")
plt.ylabel("Tiempo")
plt.title("Relación solicitudes vs tiempo")
plt.show()

# 📊 Regresión (queda más pro)
plt.figure()
sns.regplot(x=x, y=y)
plt.title("Tendencia lineal")
plt.show()

# Pearson
r, p_value = stats.pearsonr(x, y)
p_value_one_tail = p_value / 2

print("r:", r)
print("p:", p_value_one_tail)
```
